```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
plt.style.use('ggplot')
% matplotlib inline
```

```python
df = pd.read_csv('Dogs of NYC - WNYC.csv')
```

```python
df.head()
```

```python
#get rid of some duplicates by making everything lower case
df.dog_name = df.dog_name.str.lower()
```

```python
#check what names are most common
print df.dog_name.value_counts()[:10]
df.dog_name.nunique()
```

```python
#plot the top five breeds of dogs
df.breed.value_counts()[:5].plot(kind='bar')
```

```python
#quick comparison of what breeds are trained vs. not
pd.crosstab(df.breed,df.guard_or_trained).sort_values('Yes', ascending=False)
```
